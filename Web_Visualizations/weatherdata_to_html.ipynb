```python
import os
import pandas as pd
import csv
```

```python
# Store filepath in a variable
weather = "Resources/Weather_Data.csv"
```

```python
weather_df = pd.read_csv(weather)
```

```python
weather_df.head()
```

```python
wet_df = weather_df[["City", "Country", "Latitude", "Longitude", "Max Temp", "Humidity", "Cloudiness", "Wind Speed"]]
```

```python
wet_df.head()
```

```python
wet_df.count()
```

```python
wet2_df = wet_df.dropna(how='any')
```

```python
wet2_df.count()
```

```python
wet3_df=wet2_df.sort_values("City", ascending=True)
```

```python
wet3_df.head(30)
```

```python
new_index = wet3_df.reset_index(drop=True)
```

```python
new_index.head(30)
```

```python
by_latitude = new_index.sort_values("Latitude", ascending=False).reset_index(drop=True)
```

```python
by_latitude.head(30)
```

```python
by_country = new_index.sort_values("Country", ascending=True).reset_index(drop=True)
```

```python
by_country.head(30)
```

```python
by_maxtemp = new_index.sort_values("Max Temp", ascending=False).reset_index(drop=True)
```

```python
by_maxtemp.head(20)
```

```python
by_country.to_html("weather_data_by_country.html", index=True)
```

```python
by_latitude.to_html("weather_data_by_latitude.html", index=True)
```

```python
new_index.to_html("weather_data_by_city.html", index=True)
```

```python
by_maxtemp.to_html("weather_data_by_maxtemp", index=True)
```
